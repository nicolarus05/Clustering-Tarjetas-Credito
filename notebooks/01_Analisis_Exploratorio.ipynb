## ðŸ” 4. AnÃ¡lisis de Valores Nulos

Identificamos si existen datos faltantes en el dataset.

```python
# Valores nulos
print("ðŸ”Ž AnÃ¡lisis de valores nulos:\n")

nulos = df.isnull().sum()
nulos_df = pd.DataFrame({
    'Columna': nulos.index,
    'Nulos': nulos.values,
    '% Nulos': (nulos.values / len(df) * 100).round(2)
})

nulos_df = nulos_df[nulos_df['Nulos'] > 0].sort_values('Nulos', ascending=False)

if len(nulos_df) > 0:
    print("âš ï¸ Columnas con valores nulos:")
    print(nulos_df.to_string(index=False))
else:
    print("âœ… No hay valores nulos en el dataset")
```

```python
# Verificar duplicados
duplicados = df.duplicated().sum()

print(f"\nðŸ”„ Registros duplicados: {duplicados}")
if duplicados > 0:
    print(f"   Porcentaje: {(duplicados/len(df)*100):.2f}%")
else:
    print("   âœ… No hay duplicados")
```

## ðŸ“ˆ 5. DistribuciÃ³n de Variables

Visualizamos la distribuciÃ³n de las variables mÃ¡s importantes.

```python
# Seleccionar variables numÃ©ricas (excluyendo CUST_ID)
variables_numericas = df.select_dtypes(include=[np.number]).columns.tolist()
if 'CUST_ID' in variables_numericas:
    variables_numericas.remove('CUST_ID')

print(f"Variables numÃ©ricas a analizar: {len(variables_numericas)}")
```

```python
# Histogramas de variables clave
variables_visualizar = ['BALANCE', 'PURCHASES', 'CREDIT_LIMIT', 'PAYMENTS', 'CASH_ADVANCE', 'TENURE']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for idx, var in enumerate(variables_visualizar):
    axes[idx].hist(df[var].dropna(), bins=50, color='steelblue', edgecolor='black', alpha=0.7)
    axes[idx].set_title(f'{var}', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel('Valor')
    axes[idx].set_ylabel('Frecuencia')
    axes[idx].grid(True, alpha=0.3)

plt.suptitle('DistribuciÃ³n de Variables Clave', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Boxplots para detectar outliers
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for idx, var in enumerate(variables_visualizar):
    axes[idx].boxplot(df[var].dropna(), vert=True)
    axes[idx].set_title(f'{var}', fontsize=12, fontweight='bold')
    axes[idx].set_ylabel('Valor')
    axes[idx].grid(True, alpha=0.3)

plt.suptitle('Boxplots - DetecciÃ³n de Outliers', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

---
## ðŸ”— 6. AnÃ¡lisis de Correlaciones

Analizamos las relaciones entre variables mediante una matriz de correlaciÃ³n.

```python
# Matriz de correlaciÃ³n
correlacion = df[variables_numericas].corr()

plt.figure(figsize=(14, 12))
sns.heatmap(
    correlacion,
    annot=True,
    fmt='.2f',
    cmap='RdBu_r',
    center=0,
    square=True,
    linewidths=0.5,
    cbar_kws={'label': 'CorrelaciÃ³n'}
)
plt.title('Matriz de CorrelaciÃ³n entre Variables', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
# Top 10 correlaciones mÃ¡s fuertes
correlaciones_fuertes = []

for i in range(len(correlacion.columns)):
    for j in range(i+1, len(correlacion.columns)):
        correlaciones_fuertes.append({
            'Variable_1': correlacion.columns[i],
            'Variable_2': correlacion.columns[j],
            'Correlacion': correlacion.iloc[i, j]
        })

df_corr = pd.DataFrame(correlaciones_fuertes)
df_corr['Correlacion_Abs'] = df_corr['Correlacion'].abs()
df_corr_top = df_corr.sort_values('Correlacion_Abs', ascending=False).head(10)

print("\nðŸ”— Top 10 Correlaciones MÃ¡s Fuertes:\n")
print("="*80)
for idx, row in df_corr_top.iterrows():
    print(f"{row['Variable_1']:35s} â†” {row['Variable_2']:35s} : {row['Correlacion']:6.3f}")
```
